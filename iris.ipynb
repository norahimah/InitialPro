```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn import datasets
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
# Define model parameters
input_dim = 4  # Iris dataset has 4 input features
hidden_dim1 = 8  # Number of neurons in the first hidden layer
hidden_dim2 = 6  # Number of neurons in the second hidden layer
output_dim = 3  # 3 classes: Setosa, Versicolor, Virginica
num_epochs = 1000  # Number of epochs for training
batch_size = 16  # Number of samples per batch
learning_rate = 0.001  # Learning rate for training
```

```python
# Load the Iris dataset from sklearn
iris = datasets.load_iris()
data = iris.data  # Features: 4 columns
labels = iris.target  # Labels: 3 classes (Setosa, Versicolor, Virginica)

# Convert the data to a pandas DataFrame for easier handling
df = pd.DataFrame(data, columns=iris.feature_names)
df['Species'] = labels

# Dataset class for Iris dataset
class IrisDataset(Dataset):
    def __init__(self, data):
        # Features (4 columns of data)
        self.features = torch.tensor(data.drop('Species', axis=1).values, dtype=torch.float32)
        # Labels: Iris target labels (0, 1, 2 for Setosa, Versicolor, Virginica)
        self.labels = torch.tensor(data['Species'].values, dtype=torch.long)

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return self.features[idx], self.labels[idx]

# Split the dataset into training, validation, and test sets
iris_dataset = IrisDataset(df)

# Define the sizes of the splits
total_size = len(iris_dataset)
train_size = int(0.7 * total_size)  # 70% for training
val_size = int(0.1 * total_size)  # 10% for validation
test_size = total_size - train_size - val_size  # Remaining 20% for testing

# Split the dataset into train, validation, and test sets
train_data, temp_data = random_split(iris_dataset, [train_size, val_size + test_size])
val_data, test_data = random_split(temp_data, [val_size, test_size])

# Create DataLoaders
train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_data, batch_size=batch_size)
test_loader = DataLoader(test_data, batch_size=batch_size)
```

```python
# Define the neural network model for Iris dataset
class IrisNN(nn.Module):
    def __init__(self, input_dim, hidden_dim1, hidden_dim2, output_dim):
        super(IrisNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim1)  # Input layer to hidden layer 1
        self.fc2 = nn.Linear(hidden_dim1, hidden_dim2)  # Hidden layer 1 to hidden layer 2
        self.fc3 = nn.Linear(hidden_dim2, output_dim)  # Hidden layer 2 to output layer
        self.relu = nn.ReLU()
        self.softmax = nn.Softmax(dim=1)  # Softmax activation for multi-class classification

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        x = self.softmax(x)  # Output class probabilities
        return x
```

```python
## Instantiate the model
model = IrisNN(input_dim, hidden_dim1, hidden_dim2, output_dim)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()  # Cross-entropy loss for multi-class classification
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
```

```python
# Define the path to save the model
model_path = 'model/pima_model_KV.pt'
```

```python
# Function to calculate accuracy
def calculate_accuracy(outputs, labels):
    _, predicted = torch.max(outputs, 1)
    correct = (predicted == labels).sum().item()
    accuracy = correct / labels.size(0)
    return accuracy
```

```python
# Initialize lists to store loss and accuracy
epoch_losses = []
epoch_accuracies = []
val_accuracies = []
val_losses = []
best_val_accuracy = 0.0
model_path = "best_model.pth"  # Path to save the best model

# Training loop
model.train()  # Start Training
for epoch in range(num_epochs):
    epoch_loss = 0.0
    epoch_accuracy = 0.0
    for i, (input, labels) in enumerate(train_loader):  # Renamed 'features' to 'input'
        # Forward pass
        outputs = model(input)

        # Compute the loss
        loss = criterion(outputs, labels)

        # Backward pass and optimization
        optimizer.zero_grad()  # Clear previous gradient
        loss.backward()  # Compute gradient from current loss
        optimizer.step()  # Update model parameter

        # Calculate batch accuracy
        accuracy = calculate_accuracy(outputs, labels)

        epoch_loss += loss.item()
        epoch_accuracy += accuracy

    epoch_loss /= len(train_loader)
    epoch_accuracy /= len(train_loader)

    epoch_losses.append(epoch_loss)
    epoch_accuracies.append(epoch_accuracy)

    # Validation
    val_loss = 0.0  # Initialize validation loss variable
    val_accuracy = 0.0
    model.eval()  # Set the model to evaluation mode
    with torch.no_grad():
        for input_val, labels_val in val_loader:
            outputs_val = model(input_val)
            loss_val = criterion(outputs_val, labels_val)  # Compute validation loss
            accuracy_val = calculate_accuracy(outputs_val, labels_val)  # Calculate validation accuracy
            val_loss += loss_val.item()  # Accumulate validation loss
            val_accuracy += accuracy_val
        val_loss /= len(val_loader)  # Compute average validation loss
        val_losses.append(val_loss)  # Store validation loss
        val_accuracy /= len(val_loader)  # Average validation accuracy
        val_accuracies.append(val_accuracy)
    model.train()  # Set the model back to training mode

    # Save model if validation accuracy improves
    if val_accuracy > best_val_accuracy:
        best_val_accuracy = val_accuracy
        best_model_params = model.state_dict()
        torch.save(best_model_params, model_path)

    # Print statistics
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}, Accuracy: {epoch_accuracy:.4f}, '
          f'Validation Loss: {val_loss:.4f}, Validation Accuracy: {val_accuracy:.4f}')

# Print final results
avg_loss = sum(epoch_losses) / len(epoch_losses)
print(f"Training complete. Average Loss: {avg_loss:.4f}, Best Validation Accuracy: {best_val_accuracy:.4f}")
```
